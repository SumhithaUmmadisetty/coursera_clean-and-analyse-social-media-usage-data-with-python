```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpy
import seaborn as sb
import random
cat=['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family','Health']
n=1000
data={ 'date': pd.date_range('2021-01-01', periods=n),
      'category': [random.choice(cat) for _ in range(n)],
      'likes': np.random.randint(0, 10000, size=n)}
df= pd.DataFrame(data)
print(df)
print(df.head())
print(df.info())
print(df.describe())
print(df['category'].value_counts())
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df['date'] = pd.to_datetime(df['date'])
df['likes'] = df['likes'].astype(int)
sb.histplot(df['likes'])
mpy.show()
sb.boxplot(x='category', y='likes', data=df)
mpy.show()
print(df['likes'].mean())
print(df.groupby('category')['likes'].mean())
```
